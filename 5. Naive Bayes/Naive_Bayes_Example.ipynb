# Solution   

p(yes | sunny , hot , high , weak)  =  p(sunny | yes)  *  p(hot | yes)  *  p(high | yes)  *  p(weak | yes)  *  p(yes)  

  <br>
p( no | sunny , hot , high , weak)  =  p(sunny | no)  *  p(hot | no)  *  p(high | no)  *  p(weak | no)  *  p(no)

now compare both results and return maximum one
