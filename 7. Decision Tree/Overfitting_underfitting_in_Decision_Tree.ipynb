```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.tree import DecisionTreeClassifier
```

```python
pd.read_csv('/content/Social_Network_Ads.csv').head(2)
```

```python
def analyzer(max_depth = None):
  df = pd.read_csv('/content/Social_Network_Ads.csv')
  x = df.iloc[:,2:4].values
  y = df.iloc[: , -1].values

  clf = DecisionTreeClassifier(max_depth = max_depth)
  clf.fit(x,y)

  a = np.arange(start = x[:,0].min() - 1 , stop = x[: , 0].max() + 1 , step = 0.1)
  b = np.arange(start = x[: , 1].min() - 1 , stop = x[: , 1].max() + 1 , step = 100)

  xx ,yy = np.meshgrid(a , b )

  input_array =  np.array([xx.ravel() , yy.ravel()]).T

  y_pred = clf.predict(input_array)

  plt.contourf(xx,yy,y_pred.reshape(xx.shape) , alpha = 0.5 , cmap = 'jet')
  plt.scatter(x[:,0], x[:,1] , c = y)
```

```python
analyzer(max_depth = None)
```

```python
analyzer(max_depth = 1)
```

```python
analyzer(max_depth = 2)
```

```python
analyzer(max_depth = 3)
```

```python
analyzer(max_depth = 4)
```

```python
analyzer(max_depth = 5)
```
