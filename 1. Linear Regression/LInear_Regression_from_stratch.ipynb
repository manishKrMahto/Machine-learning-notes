```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
#Linear regression model class

class LinearRegression:
  def __init__(self):
    self.b = None
    self.m = None

  def fit(self , x_train , y_train):
    num = 0
    deno = 0

    # x_train == 2d array
    # y_train == series
    for i in range(x_train.shape[0]):
      num += (x_train[i] - x_train.mean() ) * (y_train.iloc[i] - y_train.mean() )
      deno  += (x_train[i] - x_train.mean() ) ** 2
      # print(x_train[i], y_train.iloc[i])

    self.m = num/deno

    self.b = y_train.mean() - self.m * x_train.mean()

    # print(f'm : {self.m} \nb : {self.b}')


  def predict (self,x_test):
    return (self.m * x_test + self.b)
```

```python
df = pd.read_csv('/content/placement_linear_regression.csv')
df.shape
```

```python
df.head(3)
```

```python
x_train , x_test , y_train , y_test = train_test_split(df['cgpa'] , df['package'] , test_size = 0.2 , random_state=2)
```

```python
x_train = x_train.values.reshape(x_train.shape[0],1)
x_test = x_test.values.reshape(x_test.shape[0], 1)
```

```python
lr = LinearRegression()
```

```python
lr.fit(x_train , y_train)
```

```python
x_test.shape
```

```python
# for clear understaing i'm creating a dataframe
pd.DataFrame (
    {
        'predicted': lr.predict(x_test).round(2).reshape(x_test.shape[0]),
        'acctual': y_test.values,
        'differece': (y_test.values - lr.predict(x_test).round(2).reshape(x_test.shape[0]))
    }
)
```
