```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_regression
from sklearn.linear_model import LinearRegression
```

- **n_samples=4:** This sets the number of samples (data points) to be generated. In this case, 4 samples will be created. <br> <br>

- **n_features=1:** This specifies the number of features (independent variables) in the dataset. Here, it is set to 1, meaning each sample will have a single feature.<br> <br>

-** n_informative=1:** This sets the number of informative features. An informative feature is one that is actually used to build the output variable. Since it's set to 1, it means the single feature we have is informative. <br><br>

- **n_targets=1:** This specifies the number of target variables (dependent variables) to be generated. In this case, there is only one target variable.
<br><br>
- **noise=80:** This parameter adds Gaussian noise to the output. The noise level is set to 80, which is relatively high and will add significant variability to the target variable.<br><br>

- **random_state=13:** This sets the seed for the random number generator. Setting the random state ensures that the same random numbers are generated every time you run the code, making the results reproducible.

```python
X , y = make_regression (n_samples = 4 , n_features = 1 , n_informative = 1 , n_targets = 1 , noise = 80 , random_state = 13 )
```

```python
print(X.shape)
print(y.shape)
```

```python
print(X)
print(y)
```

```python
# ploting graph for X and y
plt.scatter(X,y)
plt.title('X v/s y')
plt.xlabel('X')
plt.ylabel('y')
```

```python
# applying OLS (ordinary least square)
lr = LinearRegression()
```

```python
lr.fit(X ,y)
```

```python
lr.coef_  # m
```

```python
lr.intercept_   # b
```

```python
# ploting best fit line
plt.scatter(X,y)
plt.plot (X , lr.predict(X) , c = 'red')
plt.title('Best Fit Line')
plt.xlabel('X')
plt.ylabel('y')
```

# Gradient Descent
let's apply Gradient Descent assuming slop is constant m = 78.35 <br>and lest's assume th estarting value for intercept b = 0

```python
# you can change b into any number no resttriction

# y = mx + b

y_pred =  78.35 * X + 0
y_pred
```

```python
# ploting best fit line
plt.scatter(X,y)
plt.plot (X , lr.predict(X) , c = 'red' , label = 'OLS')
plt.plot(X , y_pred , c = 'green' , label ='GD  b = 0')
plt.title('Best Fit Line')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
plt.show()
```

Good your choose value of b is very close let's push it( **green** ) into the right line( **red** )

```python
X
```

```python
X.ravel()
```

`X.ravel() : ` is used to change a 2-dimensional array or a
multi-dimensional array into a contiguous flattened array

```python
# Iteration 1

m = 78.35
b = 0   # b_old
learning_rate = 0.1


# it comes from differentiation of loss_function or Error function
loss_slop = -2 * np.sum( y - m * X.ravel() - b)
loss_slop
```

```python
# let's calculate the step_size

step_size = learning_rate * loss_slop
step_size
```

```python
# calculating the new intercept (b_new)

b = b - step_size
b
```

```python
# y = mx+b
y_pred1 = (m * X) + b

# ploting best fit line
plt.scatter(X,y)
plt.plot (X , lr.predict(X) , c = 'red' , label = 'OLS')
plt.plot(X , y_pred , c = 'green' , label ='GD  b = 0')
plt.plot(X , y_pred1 , c = 'orange' , label ='GD  b = 20.927763408209216')
plt.title('Best Fit Line')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
plt.show()
```

```python
# Iteration 2

loss_slop = - 2 * np.sum(y - m * X.ravel() - b)
loss_slop
```

```python
step_size = learning_rate * loss_slop
step_size
```

```python
b = b - step_size
b
```

```python
# y = mx+b
y_pred2 = (m * X) + b

# ploting best fit line
plt.scatter(X,y)
plt.plot (X , lr.predict(X) , c = 'red' , label = 'OLS')
plt.plot(X , y_pred , c = 'green' , label ='GD  b = 0')
plt.plot(X , y_pred1 , c = 'orange' , label ='GD  b = 20.927763408209216')
plt.plot(X , y_pred2 , c = 'black' , label ='GD  b = 25.11331608985106')
plt.title('Best Fit Line')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
plt.show()
```

```python
# Iteration 3

loss_slop = - 2 * np.sum ( y - m * X.ravel() - b )

step_size = learning_rate * loss_slop

b = b - step_size
b
```

```python
# y = mx+b
y_pred3 = (m * X) + b

# ploting best fit line
plt.scatter(X,y)
plt.plot (X , lr.predict(X) , c = 'red' , label = 'OLS')
plt.plot(X , y_pred , c = 'green' , label ='GD  b = 0')
plt.plot(X , y_pred1 , c = 'orange' , label ='GD  b = 20.927763408209216')
plt.plot(X , y_pred2 , c = 'yellow' , label ='GD  b = 25.11331608985106')
plt.plot(X , y_pred3 , c = 'black' , label ='GD  b = 25.95042662617943')
plt.title('Best Fit Line')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
plt.show()
```

red and black overlaps

```python
# let's make it in a loop

b = -100  # initially random_value
m = 78.35  # known
learning_rate  = 0.01  # b_new is directly propostional to the value jumps

epochs = 100  # how many time the loop will run (randomally or guess)

for i in range(epochs):
  loss_slop = -2 * np.sum(y - m * X.ravel() - b )

  step_size = learning_rate * loss_slop

  b = b - step_size

  y_predict = m * X + b

  plt.plot(X , y_predict)


plt.scatter(X,y , c = 'blue')
plt.plot (X , lr.predict(X) , c = 'red' , label = 'OLS')
plt.title('Best Fit line OLS and GD ')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
```
