```python
import numpy as np
import pandas as pd

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score , confusion_matrix , precision_score , recall_score , f1_score
```

```python
df = pd.read_csv('/content/drive/MyDrive/Datasets/Digit Recognizer/train.csv')
df.shape
```

```python
df.head()
```

```python
x = df.iloc[: , 1:]
y = df.iloc[:,0]

x.shape , y.shape
```

```python
x
```

```python
y
```

```python
x_train , x_test , y_train ,y_test = train_test_split(x,y,test_size = 0.2,random_state = 2)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
clf1 = LogisticRegression()
clf2 = DecisionTreeClassifier()
```

```python
clf1.fit(x_train, y_train)
clf2.fit(x_train, y_train)
```

```python
y_pred1 =  clf1.predict(x_test)
y_pred2 = clf2.predict(x_test)
```

# accuracy score

```python
print("Accuracy of Logistic Regression",accuracy_score(y_test,y_pred1))
print("Accuracy of Decision Trees",accuracy_score(y_test,y_pred2))
```

# Confusion matrix

```python
print("Logistic Regression Confusion Matrix\n")
pd.DataFrame(confusion_matrix(y_test,y_pred1),columns=list(range(0,10)))
```

```python
print("Decision Tree Confusion Matrix\n")
pd.DataFrame(confusion_matrix(y_test,y_pred2),columns=list(range(0,10)))
```

```python
print('Logistic Regression')
print('precision score ' , precision_score(y_test,y_pred1,average=None))
print('recall score ' , recall_score(y_test,y_pred1,average=None))
print('f1 score ' , f1_score(y_test , y_pred1 , average = None))
```

```python
print('Logistic Regression')
print('precision score ' , precision_score(y_test,y_pred2,average=None))
print('recall score ' , recall_score(y_test,y_pred2,average=None))
print('f1 score ' , f1_score(y_test , y_pred2 , average = None))
```

# classification report is equevalent to this method (accuracy , macro avg , weighted avg , f1 score)

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred1))
```
