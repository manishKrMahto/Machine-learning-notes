```python
import numpy as np
import pandas as pd

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score , confusion_matrix
from sklearn.metrics import recall_score,precision_score,f1_score
```

```python
heart = pd.read_csv('heart.csv')
heart.shape
```

```python
heart.head()
```

```python
x = heart.iloc[:,:-1]
y = heart.iloc[:,-1]

x.shape , y.shape
```

```python
x
```

```python
y
```

```python
x_train , x_test , y_train , y_test = train_test_split(x,y,test_size = 0.2 , random_state = 2)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
clf1 = LogisticRegression()
clf2 = DecisionTreeClassifier()
```

```python
clf1.fit(x_train,y_train)
clf2.fit(x_train ,y_train)
```

```python
y_pred1 = clf1.predict(x_test)
y_pred2 = clf2.predict(x_test)
```

# Accuracy Score

```python
print('accuracy score for LOGISTIC REGRESSION --> ', accuracy_score(y_test , y_pred1))
print('accuracy score for DECISION TREE --> ', accuracy_score(y_test , y_pred2))
```

# Confusion metrix

```python
print('confusion matrix for LOGISTIC REGRESSION ' )
pd.DataFrame(confusion_matrix(y_test , y_pred1) , columns = [0,1])   # columns = list (range(0,2))
```

```python
print('confusion matrix for DECISION TREE' )
pd.DataFrame(confusion_matrix(y_test , y_pred2) ,  columns = list (range(0,2)) )
```

```python
# better understanding
result = pd.DataFrame()
result['Actual Label'] = y_test
result['Logistic Regression Prediction'] = y_pred1
result['Decision Tree Prediction'] = y_pred2

result.head(10)
```

# precision , recall , f1_score

```python
print("For Logistic regression Model")
print("-"*50)
cdf = pd.DataFrame(confusion_matrix(y_test,y_pred1),columns=list(range(0,2)))
print(cdf)
print("-"*50)
print("Precision - ",precision_score(y_test,y_pred1))
print("Recall - ",recall_score(y_test,y_pred1))
print("F1 score - ",f1_score(y_test,y_pred1))
```

```python
# precision , recall is calculated for all the classes but we get class 1 result in binary classification
precision_score(y_test,y_pred1,average=None)
```

```python
recall_score(y_test,y_pred1,average=None)
```

```python
print("For Decision Tree Model")
print("-"*50)
cdf = pd.DataFrame(confusion_matrix(y_test,y_pred2),columns=list(range(0,2)))
print(cdf)
print("-"*50)
print("Precision - ",precision_score(y_test,y_pred2))
print("Recall - ",recall_score(y_test,y_pred2))
print("F1 score - ",f1_score(y_test,y_pred2))
```

```python
precision_score(y_test,y_pred2,average=None)
```
