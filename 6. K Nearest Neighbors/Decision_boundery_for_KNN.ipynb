# 1.Decision Boundery for KNN  
### 1. What is Decision Boundary   
in a classification problem with two or more classes. a decision boundary or decision surface is hypersurface that partitions the underlying vector space into two or more sets. one for each class. The classifier will classify all the points on one side of the decision as belonging to the class and all those on the other side as belonging to the other class

<b> <h3> Important Points </h3> </b>
1. we can draw decision boundary for all the classification algorithms including Neural Networks .  
2. Decision boundary can be both linear(as in the case of SVM) or non-linear (as in the case of Decision tree classifier or KNN)  
3. Decision Boundries are not always clear cut. That is the transition from one class in the freature space to another is not discontinuous. but gradual. This effect is common in fuzzy logic based classification algorithms. where membership in one class or another is ambigyous.  
4. For higher dimension problems the decision boundary acts as a hyperplane (for linear ones).

in mathematics. a Voronoi Diagram is partitioning of a plane into regions based on distance to points in a specific subset of the plane

<b> <h2> Steps to plot Decision boundary for KNN(Assuming 2 input cols) </h2></b>  
1. Train the classifier on the training set.  
2. Create a uniform grid (with the help of NUMPY meshgrids) of points that densely cover the region of input space containing the training set.  
3. classify each point on the grid. Store the results in an array A, contains the predicted class for the point at row i , column j on the grid.  
4. PLot the array as an image, where each pixel corresponds to a grid point and it's color represents the predicted class. The Decision boundary can be seen as <b color = 'red'> contour plot </b> where the imgae changes color.  
5. Finally print out training data with their respective color on the same <b color = 'red'> contour plot </b>.

# Effect of k on Decision Boundary

```python
knn = KNeighborsClassifier(n_neighbors = 17)
knn.fit(x_train , y_train)
y_pred = knn.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix

c = confusion_matrix(y_test , y_pred)
c
```

```python
(c[0][1] + c[1][0])/c.sum()
```

<p color = 'green'>Major diagonal element bata raha h kitna sahi prediction h </p>  
<p color = 'red' > non major diagonal elements bata raha h ki kitna bar wrong prediction kiya h </p>

```python
# Effect of k on training error rate and validation error rate

error_train = []
error_test = []

lower = 1
upper = 26

for i in range(lower , upper):
  knn = KNeighborsClassifier(n_neighbors = i)
  knn.fit(x_train  , y_train)
  x = confusion_matrix(y_train , knn.predict(x_train))
  y = confusion_matrix(y_test , knn.predict(x_test))

  # finding percent error (wrong prediction) in training data
  error_train.append((x[0][1] + x[1][0]) / x.sum())
  # finding percent error (wrong prediction) in test data
  error_test.append((y[0][1] + y[1][0]) / y.sum())
```

```python
plt.plot(range(lower , upper) , error_train , label = 'training error rate')
plt.plot (range(lower , upper) , error_test , label = 'test/validation error rate')
plt.xlabel('K values')
plt.ylabel('Error')
plt.legend()
```

## KEffect

```python
def keffect(dataset , k):
  df = pd.read_csv(dataset)
  x = df.iloc[: , :2].values
  y = df.iloc[: ,-1].values
  print(df)

  print(x.shape)
  print(y.shape)
  knn = KNeighborsClassifier(n_neighbors = k)
  knn.fit(x,y)

  a = np.arange(start = x[: , 0].min() - 1 , stop = x[: , 0].max() + 1 , step = 0.01)
  b = np.arange(start = x[: , 1].min() - 1 , stop = x[: , 1].max() + 1 , step = 0.01)

  print(a.shape)
  print(b.shape)
  xx , yy = np.meshgrid(a , b )

  input_array = np.array([xx.ravel() , yy.ravel()]).T
  print(input_array.shape)

  y_pred = knn.predict(input_array)

  plt.contourf(xx, yy , y_pred.reshape(xx.shape) , cmap = 'winter' , alpha = 0.5)
  plt.scatter(x[:, 0] , x[: ,1] , c = y , cmap = 'jet')
```

```python
keffect('/content/1.ushape.csv', 1)
```
