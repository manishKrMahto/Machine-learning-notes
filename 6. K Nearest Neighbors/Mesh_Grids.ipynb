```python
x = np.array([1,2,3,4])
y = np.array([5,6,7])

xx , yy = np.meshgrid(x,y)
xx.shape        # is equal to yy.shape
```

## Purpose of Meshgrid  
meshgrid is very useful to evaluate functions on a grid . we can apply any function to the points of a meshgrid to plot a function

### ploting a function using meshgrid

```python
x = np.linspace (-40 , 40 , 100)
y = np.linspace (-50 , 50 , 90)

xx , yy = np.meshgrid(x,y)
xx.shape      # is equal to yy.shape
```

<p> z = xx<sup>2</sup> + yx<sup>2</sup>

```python
z = (xx ** 2) + (yy ** 2)
```

```python
# ploting a filled contour plot
plt.contourf(xx,yy,z)
```

```python
# with a random values
r = np.random.random((90,100))

plt.contourf(xx,yy , r)
```
